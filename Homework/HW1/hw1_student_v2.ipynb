The behavior of sample distributions within the theoretical sigma contours is significantly influenced by the noise characteristics of the sensor measurements:

1. Range and Bearing Noise:

2. Balanced Noise:
If the noise in range and bearing measurements is more balanced or relatively small, the non-linear effects of the transformation are less pronounced. Consequently, the distribution in Cartesian coordinates may more closely resemble a Gaussian distribution. This leads to a closer match between the actual count of samples within the sigma contours and the theoretically expected values.

Consequences to State Estimation

The assumption of Gaussian noise is central to algorithms like the Kalman Filter. Deviations from this assumption can have significant implications:

1. Non-Gaussian Distributions:
In cases where sensor noise leads to non-Gaussian distributions, particularly after non-linear transformations, algorithms that assume Gaussian noise (such as the Kalman Filter) might yield suboptimal results. This is because these algorithms are optimized under the premise of Gaussian noise distributions.
2. SLAM and Navigation Accuracy:
In SLAM (Simultaneous Localization and Mapping) or robotic navigation, inaccuracies in mapping and localization can arise if the non-linearity of sensor noise is underestimated. Accurate predictions and decisions depend on the reliability of these estimations.
3. Adaptive Algorithms:
More sophisticated algorithms like Particle Filters or Unscented Kalman Filters may be required to handle significant non-linearities and non-Gaussian distributions effectively. These algorithms are designed to accommodate a wider range of noise distributions.
4. Sensor Fusion:
Combining data from multiple sensors, each with different noise characteristics, can mitigate the effects of non-linearities. This approach can leverage the strengths of each sensor to achieve a more accurate and reliable estimation.

### Problem 5. Estimation

Assume that we want to estimate an unobserved population parameter $\theta$ on the basis of observations $x$. Let $f$ be the sampling distribution of $x$ so that $f(x|\theta)$ is the probability of $x$ when the underlying population parameter is $\theta$. The function $L(\theta) = f(x|\theta)$ when viewed as a function of the parameter $\theta$ is called the likelihood function or just the likelihood.
For example, if $x$ follows a Gaussian distribution, we will have $\theta = (\mu, \sigma^2)$ and $$\mu, \sigma \mapsto f(x|\mu,\sigma^2) = \frac{1}{\sigma \sqrt{2\pi}}\exp(-\frac{1}{2}\frac{(x-\mu)^2}{\sigma^2})$$ is the likelihood. 

**Maximum Likelihood Estimator (MLE):** The maximum likelihood method maximizes the likelihood function, leading to the MLE
\begin{equation} 
    \nonumber \hat{\theta}_{MLE} = \text{argmax}_\theta L(\theta) = \text{argmax}_\theta f(x|\theta).
\end{equation}

**Maximum A Posteriori (MAP) Estimator:** In the Bayesian framework, one can place a prior distribution over the parameter, i.e., $g(\theta)$. Then, the MAP estimator maximizes the posterior probability density function $f(\theta|x)$ as follows.
\begin{equation}\nonumber
    \hat{\theta}_{MAP} = \text{argmax}_\theta f(\theta|x)
    = \text{argmax}_\theta \frac{f(x|\theta)g(\theta)}{\int_\Theta f(x|\vartheta)g(\vartheta)d\vartheta}
    = \text{argmax}_\theta f(x|\theta)g(\theta),
\end{equation}
were the last equality is true because the normalization constant in the Bayes' formula is independent of $\theta$.

**Remark 1:**
Since $\log$ is a monotonic function, it is often the case that we use the logarithm of the likelihood or posterior for maximization (or negative of the logarithm for minimization).
